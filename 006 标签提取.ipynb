#### 肾柱肥大

```python
import pandas as pd
import re
```

```python
test_recode = pd.read_excel(r"C:\Users\13670\Desktop\待完成\项目4 回声tag分析\肾脏\20201202\test\check1.xlsx")
```

```python
keyword='肾柱肥大'
ORDER_CODE_list = []
desc_list = []
summ_list = []
keyword_list = []

for index, row in test_recode.iterrows():
    try:
        result_summ = re.findall(keyword, row['RAW_RESULT'])
        result_desc = re.findall(keyword, row['RESULT'])
    except:
        continue
    if result_summ != [] or result_desc != []:
        ORDER_CODE_list.append(index)
        desc_list.append(row['RESULT'])
        summ_list.append(row['RAW_RESULT'])
        if result_summ != []:
            keyword_list.append(result_summ)
        else:
            keyword_list.append(result_desc)
        
df = pd.DataFrame({"订单号":ORDER_CODE_list,"报告描述":desc_list,"报告小结":summ_list,"提取词":keyword_list})
df
```
